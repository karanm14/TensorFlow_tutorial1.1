# So far we learnt about constants, now lets learn about variables!
we will learn how to create a variable and how to initialize it randomly

```python
rand_t = tf.random_uniform([50,50],0,10,seed=12)
t1 = tf.Variable(rand_t)
t2 = tf.Variable(rand_t)

a=tf.subtract(t1,t2)
initial_op = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(initial_op)
    #always remember to initalise the variable
    print(sess.run(a))

```

```python
weights = tf.Variable(tf.random_normal([100,100],stddev=2.0), name="w1")
bias = tf.Variable(tf.zeros(100), name="b1")
initial_op = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(initial_op)
    print(weights.shape)
    print(bias.shape)
    print(sess.run(weights+bias))
    

```

# Now lets go onto know about placeholders
this is the most important of 3 because this is how you will input data

```python
x=tf.placeholder("float")
y =2*x
data = tf.random_uniform([3,3],10)


with tf.Session() as sess:
    x_data=sess.run(data)
    print(sess.run(y,feed_dict={x:x_data}))
```

```python
w=tf.Variable(0,dtype=tf.float32)
#cost=tf.add(tf.add(w**2,tf.multiply(-10.,w)),25)
cost = w**2 -10*w +25
train=tf.train.GradientDescentOptimizer(0.01).minimize(cost)

init=tf.global_variables_initializer()
session = tf.Session()
session.run(init)
print(session.run(w))
```

```python

session.run(train)
print(session.run(w))
```

```python

for i in range(1000):
    session.run(train)
print(session.run(w))

```

```python
coefficients = np.array([[1.],[-20.],[100.]])
w=tf.Variable(0,dtype=tf.float32)
x=tf.placeholder(tf.float32,[3,1])
cost=x[0][0]*w**2 + x[1][0]*w + x[2][0]
train=tf.train.GradientDescentOptimizer(0.01).minimize(cost)

init=tf.global_variables_initializer()
session = tf.Session()
session.run(init)
print(session.run(w))

```

```python
coefficients[1][0]
```

```python

session.run(train, feed_dict={x:coefficients})
print(session.run(w))

```

```python
for i in range(1000):
    session.run(train, feed_dict={x:coefficients})#get your training data into the cost function with this syntax
print(session.run(w))
```
